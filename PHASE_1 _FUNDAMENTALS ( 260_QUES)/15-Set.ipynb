#### 6. Find the Union of two sets.


```python
# write your code here 
s1 = {1,2,3,4,5}
s2 = {10,11,12,13}
print(f"Union Set = {s1.union(s2)}")
```

#### 7. Find the Intersection of two sets.


```python
# write your code here 
s1 = {1,2,3,4,5}
s2 = {10, 1,2,4,13}
print(f"Intersection Set = {s1.intersection(s2)}")
```

#### 8. Find the Difference (A - B) of two sets.


```python
# write your code here 
A = {1,2,3,4,5}
B = {10, 1,2,4,13}
print(f"Intersection Set = {A.difference(B)}")  # A - B 
```

#### 9. Find the Symmetric Difference of two sets.


```python
# write your code here 
A = {1,2,3,4,5}
B = {10, 1,2,4,13}
print(f"Symmetric Difference  = {A.symmetric_difference(B)}")  # A - B union B - A  
```

#### 11. Check if two sets are disjoint (no common elements).


```python
# write your code here 
A = {1,2,3,4,5}
B = {e for e in range(1,11)}
if A.intersection(B) :
 print("Set A and B are not disjoint.")
else :
 print("Set A and B are disjoint.")
```
