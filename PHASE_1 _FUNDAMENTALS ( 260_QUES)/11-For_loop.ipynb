#### 1. Write a python script to print each character of a string with its corresponding Unicode.

```python
# write your code here
s = input("Enter a string: ")

for e in s:
  print(e, "->", ord(e))
```

#### 2. Write a Python script to print only vowels of the given string

```python
# write your code here
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
found = 0

for e in s:
  if e in vowels :
    print(e, end=' ')
    found = 1
    
if not found :
  print(f"No vowels found in {s}")  
```

#### 3. Write a Python script to count occurrence of spaces in a given string.

- you can do -> e == ' '
- s.count(' ')
- isspace() --> this checks white space , new line , tab

```python
# write your code here
s = input("Enter a string: ")
count = 0

for e in s:
  if e.isspace():
    count +=1
print(count)    
```

#### 4. Write a Python script to print unique digits of a given integer.

```python
# write your code here
num =  input("Enter a number: ") 
temp = ""
for e in num:
  if e not in temp:
    print(e, end=" ")
    temp += e
```

In Python, a Set is a built-in data structure designed specifically to store unique items

```python
num = input("Enter a number: ")

# A set automatically removes duplicates
unique_digits = set(num)

for digit in unique_digits:
    print(digit)
```

Method 1: The "Math" Way
Using num //= 10 is the classic way to solve this in languages like C or Java.

Pros: It works specifically on the numerical value.

The Zero Catch: If you enter 0, the loop while num: won't run at all, and your count will stay 0 (even though 0 is technically one digit).

Method 2: The "String" Way
Using a for loop on a string input is the most common way in Python.

Pros: It is very fast and handles the digit 0 correctly (count will be 1).

Cons: If the user enters a negative sign (e.g., -123), your loop will count the - as a digit, giving you 4 instead of 3.

```python
num = int(input("Enter a number : "))

count  = len(str(abs(num)))
print("Count =",count)
```

#### 15. Write a python script to display all prime numbers within a range. Take start end values form user

`We can use else statement with loop But that else statement will only be executed when the loop do not terminates Abruptly.`

```python
# write your code here
start = int(input("Enter start value = "))
end = int( input("Enter end value = "))

print(f"Prime numbers in range({start}-{end}) :")
for e in range(start, end+1) :
  for i in range(2,e):   
    if e % i == 0 :
      break
  else :
    print(e, end=" ")
```
