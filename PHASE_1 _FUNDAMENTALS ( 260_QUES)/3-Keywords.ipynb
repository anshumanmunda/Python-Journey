5. What is the use of del keyword?
  

Ans: When we create variable like x = 5.
  On writing this a variable is created named x inside name space/stack.
  
  All the objects are created on heap space (private heap space). So inside this heap space. A object is created of int class and that object contain a value = 5. And the ID or the reference number of this object is stored inside variable x. so x refers to that object x -> 5

  If we want to delet the object of heap space which contain 5, So that is not possible. In Python we have automatic memory management and the memory of this object will be released automatically.

  So when we write del x ---> this means we are deleting the variable (reference variable) x from namespace.

  Now no one is pointing to the object which was created. So that object will become a garbage block. So in such situations Python calls the garbage collector and memory of this garbage block will be removed automatically.



  Final answer: del keyword is not used to delete the object, it is used to remove or delete the reference variable which is created in the namespace.

  So when  we write 'del x' that means This reference variable x (which is pointing to some object) will be removed from the namespace and when we try to print the value of x there will be an error.
