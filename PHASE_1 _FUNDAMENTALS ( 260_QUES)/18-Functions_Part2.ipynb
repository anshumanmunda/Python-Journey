#### 1. Write a function filter_even(numbers) that takes a list of integers and returns a new list containing only the even numbers.
> **Logic:** This practice focuses on passing a List as an argument and returning a new List based on conditional logic.


```python
# write your code here 
def filter_even(l):
  new_lst = [ e for e in l if e & 1 == 0 ]
  return new_lst

lst = [10,23,4,6,8,99]
new_lst = filter_even(lst)
print(new_lst)
```

### 7. Create a function square_elements(nums) that takes a list, squares every number in-place (modifies the original list), and returns nothing.
> **Logic:** Understand the concept of 'Mutablity'â€”how functions can modify the original data structure without a return statement.


```python
# write your code here 
def square_elements(nums) :
  i = 0
  for e in nums:
    nums[i] =  e  ** 2
    i += 1
    
l = [e for e in range(1,11)]
square_elements(l)
print(l)
```

### 8. Write a function that takes a full name (string) and returns the first name, middle name, and last name as separate values (handle cases with no middle name by returning None).
> **Logic:** Practice string splitting and conditional packing for multiple return values.


```python
# write your code here 

def full_name(name):
  l = name.split(' ')
  length = len(l)
  if length < 2 :
    print("Please Enter full name")
    return 
  
  elif length > 2 :
    f_name = l[0]
    m_name = l[1]
    l_name = l[2]
    return {'First Name':f_name,'Middle Name': m_name, 'Last Name': l_name}
  else: 
    f_name = l[0]
    m_name = None
    l_name = l[1]
    return {'First Name':f_name,'Middle Name': m_name, 'Last Name': l_name}

name = input("Enter your full name: ")
print(full_name(name))
```

### 9. Write a function total_marks(student_data) where student_data is a list of lists (e.g., [[90, 80], [70, 85]]). Return a new list containing the sum of each sub-list.
> **Logic:** Work with nested data structures and use loops within functions to process sub-collections.


```python
# write your code here 
def total_marks(student_data):
  new_lst = list()
  for e in student_data :
    s = 0
    s += sum(e)
    new_lst.append(s)
  return new_lst  

marks_list = [[90, 80], [70, 85], [90,88,68]]
print(total_marks(marks_list))
```

### 10. Write a function char_count(s) that takes a string and returns a dictionary representing the frequency of each character (e.g., 'apple' -> {'a': 1, 'p': 2, 'l': 1, 'e': 1}).
> **Logic:** The classic frequency-map problem: building a dictionary dynamically while iterating through a string.


```python
# write your code here 
def char_count(s) :
  counted = ""
  freq = {}
  for e in s:
    if e not in counted:
      freq[e] = s.count(e)
      counted  += e 
  return freq    
    
s = 'apple'
print(char_count(s))
```
