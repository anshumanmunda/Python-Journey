#### 10. Reverse a string using a loop (not slicing).


```python
# write your code here 
s = 'anshuman'
rev = ""
i = len(s) - 1

while i >= 0 :
  rev += s[i]
  i -= 1 
print(rev)  
```

#### 11. Check if a string is a palindrome (reads the same forward and backward).


```python
# write your code here 
s = 'anshu'.lower().remove(" ","")
if s[:] == s[::-1] :
  print("Palindrome")
else :
  print("Not Palindrome")
```

#### 12. Write a program to count the frequency of each character in a string and print.


```python
# write your code here 
s = input("Enter a string: ")
seen = ""

for e in s :
  if e not in seen :
    count =  s.count(e)
    seen += e
    print(f"{e} -> {count}")
```

#### 16. Find the longest word in a given sentence.


```python
# write your code here 
sen = input("Enter a sentence: ") 
word = sen.split()
longest_word = word[0]

for e in word :
  if len(e) > len(longest_word) : 
    longest_word = e
print(f"Longest word: {longest_word}")  
```

#### 17. Write a program to remove all special characters (like @, #, $, %) from a string.


```python
# write your code here 
s  = "anshu@123!@%"
special_character = "!@#$%"
new_str = ""

for e in s :
  if e not in special_character :
    new_str += e
print(new_str)
```

#### 19. Mask a credit card number so only the last 4 digits are visible (e.g., '************4412').


```python
# write your code here 
s = input("Enter 10 digit card number: ")
last_four_digit = s[-4:]
mask = len(s) - 4
masked_num = ""
masked_num = '*' * mask + last_four_digit
print(masked_num)
```
