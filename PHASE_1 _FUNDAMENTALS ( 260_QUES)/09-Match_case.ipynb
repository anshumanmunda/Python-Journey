#### 1. Write a python script to check whether a given number is a three digit number or not.

```python
# write your code here
num = int(input("Enter a number: "))

match num:
  case num if num > 99 and num < 1000:
    print(f"{num} is a 3-digit number.") 
  case _:  
    print(f"{num} is not a 3-digit number.") 
```

#### 2. Write a python script to check whether a given number is positive, negative or zero.

```python
# write your code here.
num = int(input("Enter a number: "))

match num:
  case num if num > 0 :
    print(f"{num} is a positive number.") 
  case num if num < 0 :  
    print(f"{num} is a negative number.") 
  case _: 
    print(f"{num} is a zero.") 
```

#### 3. Write a python script to make a menu driven program in which user has to choose one of the option from four given options - 1) Odd-Even, 2) positive - Non Positive, 3) Simple Interest and 4) find roots of quadratic equation. Match and execute appropriate code on user selection.

- Use this cmath instead of 'math' for negative roots

```python
# write your code here
import cmath

print("1) Odd-Even\n2) positive - Non Positive\n3) Simple Interest\n4) find roots of quadratic equation.")
choice = int(input("Enter your choice (1-4): "))

match choice:
  case 1:
    print("-------- 1) Odd-Even --------\n")
    num = int(input("Enter a n number: "))
    if num & 1 :
      print("Odd number")
    else :  
      print("Even number")

  case 2 :
    print("-------- 2) Positive - Non Positive --------\n")
    num = int(input("Enter a n number: "))
    if num > 0 :
      print("Positive number")
    else :  
      print("Non Positive number")

  case 3 :
    print("-------- 3) Simple Intrest --------\nEnter Principal, rate and time\n")
    p, r, t = float(input("p = ")), float(input("r = ")), float(input("t = "))
    print("Simple Intreast =", (p * r * t) / 100)

  case 4 :
    print("-------- 4) Roots of quadratic Equation --------\n")
    print("Enter the coefficients of quadratic equation (ax^2 + bx + c = 0)")
    a, b, c = int(input("a = ")), int(input("b = ")), int(input("c = "))
   
    D = b**2 - 4*a*c

    root1 = (-b + cmath.sqrt(D)) / (2*a)
    root2 = (-b - cmath.sqrt(D)) / (2*a)

    if D >= 0:
        print(f"Real Roots: {root1.real} and {root2.real}")
    else:
        print(f"Complex Roots: {root1} and {root2}")

  case _:
    print("! Invalid Choice\nPlease try again")  
    
```

#### 4. Write a python script to take one data from user and evaluate the type of data. If the data is of int type then print Monday, if the data is of float type then print Tuesday, if the data is of complex type then print Wednesday, if the data is of type bool then print Thursday.

```python
# write your code here
data = eval(input("Enter a data : "))

match data:
  case data if type(data) == bool:
    print("Thrusday")
  case data if type(data) == int:
    print("Monday")
  case data if type(data) == float:
    print("Tuesday")
  case data if type(data) == complex :
    print("Wednesday")
  case _:
    print("Data is n non-prmitive type.")  
```

#### 5. Write a python script to take a string from the user. If the string is a part of "I" then print "One", if the string is a part of "love" then print "Two" and is a part of "python" then print "Three".

```python
# write your code here
s = input("Enter a string: ")

match s:
  case s if s in "I":
    print("One") 
  case s if s in "love":
    print("two") 
  case s if s in "python":
    print("three") 
```
