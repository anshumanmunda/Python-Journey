#### 5. Write a python script to print two given words in dictionary order

The Surprise: If you input word1 = "Zebra" and word2 = "apple", your code will print Zebra apple.

The Reason: In the computer's alphabet, Z is number 90, but a is number 97.

Python compares strings character by character, from left to right.

It compares the 1st character. If they are different, it decides the order right then and there.

If the 1st characters are identical, it moves to the 2nd character.

If those are identical, it moves to the 3rd, and so on.

```python

word1, word2 = input("word-1:"), input("word-2:")
if word1.lower() > word2.lower() :
  print(word2, word1)
else :
  print(word1, word2)
```

#### 9. Write a python script to check whether a given year is a leap year or not.

```python
# write your code here
year = int(input("Enter a year: "))

if year % 100 == 0 :
  if year % 400 == 0: 
    print(f"{year} is Leap year")
  else :  
    print(f"{year} is not a Leap year")

elif year % 4 == 0:
  print(f"{year} is Leap year")
else :
  print(f"{year} is not a Leap year")
```
