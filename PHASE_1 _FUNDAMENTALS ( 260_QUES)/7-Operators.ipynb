#### 3. Write a python script to swap data of two variables

```python
# Write your code here
n1, n2 = int(input("Enter 1st number: ")), int(input("Enter second number: "))

print("--------Before Swapping-------- \nnum1 = ",n1,"\nnum2 = ",n2)
temp = n1
n1 = n2
n2 = temp
print("-------- After Swapping--------\nnum1 = ",n1, "\nnum2 = ",n2)
```
