1. Write a python script to print any number and its binary equivalent.

If we want to convert a decimal -> binary we use bin() function.
bin function return str type value.

```python
n = 25
b = bin(n)
print(b,type(b))
```

2. Write a python script to store binary number 1100101 in a variable and print it in decimal format.

```python
bi = 0b1100101
print(bi)  # we can directly print in decimal formate
```

3. Write a python script to store a hexadecimal number 2F in a variable and print it in octal format.

```python
hx = 0x2F  
# print(hx) ---> this will print in decimal
oc = oct(hx)
print(oc, type(oc))
```

4. Write a python script to store an octal number 125 in a variable and print it in binary format.

```python
oc = 0o125
print(bin(oc))
```

5. Write a python script to add two numbers 25(in octal) and 39 (in hexadecimal) and display the result in binary format.

```python
n1 = 0o25  # base 8
n2 = 0x39  # base 16
sum = n1+n2  # here base 8 and base 16 are converted to base 10 and stored in sum 
print( bin(sum) )
```

SUMMARY 

1. Binary numbers --> base 2  || prefix 0b
2. Octal numbers --> base 8  || prefix 0o
3. Decimal --> base 10  ||no prefix
4. Hexadecimal numbers --> base 16  || prefix 0x/0X

CONVERSION 

a. Any num --> Binary ====> bin()
b. Any num --> Octal ====> oct()
b. Any num --> Hexadecimal ====> hex()
