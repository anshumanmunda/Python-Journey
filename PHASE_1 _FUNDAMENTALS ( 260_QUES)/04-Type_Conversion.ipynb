1. Write a python script to convert a number into str type.

```python
a = 10
print(a,type(a))

b = str(a)
print(b,type(b))
```

2. Write a python script to print Unicode of the character 'm'.
   Ans: To print Unicode of any character use ord() function

```python
print(ord('m'))
```

3. Write a python script to print character representation of a given unicode 100.

```python
n = 100
x = chr(n)
print(x)
```

4. Write a python script to convert a str type data into an int type. Also describe when a str type value is not possible to convert into an int type.

Ans: 
CASE 1:  if string variable contain number(Pure Integer) "256".

CASE 2:  if string variable contain float value "34.4"  --> ERROR by same way
This will produce error because this is not integer.

```python
a = "256"
x = int(a)
print(a, type(a))
print(x, type(x))
```

5. How to convert an integer value into a bool value?

Rule :  
0 -> False
Every non-zero -> True 

```python
n = 100
x = bool(n)
print(n, type(n))
print(x, type(x))
```
