#### 14. Write a program to remove all duplicate elements from a list.


```python
# write your code here 
l = [10, 20, 30, 10, 50, 20]

for e in l :
  count = l.count(e)
  if count >= 2 :
    l.remove(e)

print(f"Unique List = {(sorted(l))}")
```

#### 16. Write a program to find the second largest number in a list.


```python
# write your code here 
import random
l = [ random.randint(1,100) for _ in range(10) ]

largest_num = max(l)
second_largest = l[0]
for e in l :
  if e < largest_num and e >= second_largest :
    second_largest = e

print(l)
print(f"Second Largest number = {second_largest}")
```
