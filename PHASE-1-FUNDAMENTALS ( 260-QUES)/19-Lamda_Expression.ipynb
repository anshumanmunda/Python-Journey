### 1. Write a lambda expression that takes two arguments and returns their sum. Assign it to a variable and test it with numbers.
> **Logic:** Lambdas use the syntax `lambda arguments: expression`. They are often assigned to variables for simple math tasks.


```python
# write your code here
add = lambda a,b : a + b
print("Sum =",add(10,20))
```

### 2. Create a lambda function that takes a string and returns its length.
> **Logic:** Lambdas are excellent for quick string operations. Use the built-in len() function inside the expression.


```python
# write your code here 
len_str =  lambda s : len(s) 
print(len_str("Anshuman"))
```

### 3. Write a lambda expression to calculate the square of a number ($x^2$).
> **Logic:** This is the most basic use case for a lambdaâ€”performing a single mathematical operation on a single input.


```python
# write your code here 
square =  (lambda n : n ** 2) (10)
print(square)
```

### 4. Write a lambda expression that takes two numbers and returns the greater of the two using an if-else structure.


```python
# write your code here 
greater = lambda a,b : a if a > b else b
print(greater(10,20))
```

### 5. Create a lambda that takes a string and checks if the letter 'a' is present in it, returning True or False.
> **Logic:** Use the `in` operator inside your lambda for member checking.


```python
# write your code here 
check = lambda s: True if 'a' in s else False
print(check('anshu'))
```

### 6. Given a list of tuples [(1, 'c'), (3, 'a'), (2, 'b')], use a lambda as a key in the sort() method to sort the list by the second element.
> **Logic:** The `key` parameter in sorting methods is one of the most powerful places to use lambdas.


```python
# write your code here 
list_of_t = lambda l : sorted(l)

l = [(1, 'c'), (3, 'a'), (2, 'b')]
print(list_of_t(l))
```

### 7. Write a lambda expression that takes a string and returns it reversed (using slicing).
> **Logic:** Practice using string slicing `[::-1]` inside the lambda expression.


```python
# write your code here 
rev_str = lambda s : s[::-1]
print(rev_str("anshu"))
```

### 8. Create a lambda that takes two arguments, base and exponent, and returns base raised to the power of exponent.
> **Logic:** This demonstrates handling multiple arguments in a lambda definition.


```python
# write your code here 
power = lambda base,exp : base ** exp
print(power(2,10))
```

### 10. Use a lambda inside a filter() call to find numbers in a list that are both greater than 10 and divisible by 5.
> **Logic:** Combine logical operators like `and` within the lambda for complex filtering.


```python
# write your code here 
fun = lambda n : True if n > 10 and n % 5 == 0 else False 
l = [e for e in range(1,101)]

result = filter(fun, l)
print(list(result))
```
