#### 11. Find length of a tuple without using len().

#### 12. Slice to get the middle three elements.


```python
# write your code here 
t = (2, 5, 8, 5, 10,15,11,9, 14)
start = len(t) // 2  - 1
end =  start + 3
print(t[start: end])
```

#### 15. Swap two variables using tuple unpacking logic.


```python
# write your code here 
a = 10
b = 20
print(f"Before Swapping\na = {a} b = {b}")
a,b = b,a
print(f"After Swapping\na = {a} b = {b}")
```
