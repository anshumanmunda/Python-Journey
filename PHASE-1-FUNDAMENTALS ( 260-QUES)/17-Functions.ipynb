#### 9. Find max in list (no max() function).


```python
# write your code here 
def max(l) :
  max = l[0]
  for e in l:
    if e >= max :
      max = e 
  return max

l = [10, 1, 19, 50, 0]
print(f"Maximum number in list: {max(l)}")
```

#### 11. Mini Calculator function.


```python
def add(n1, n2): 
  return n1 +n2

def sub(n1, n2): 
  return n1 - n2

def mul(n1, n2): 
  return n1 * n2

def div(n1, n2): 
  return n1 / n2

def square(n): 
  return n ** 2

def cube(n): 
  return n ** 3

while True :
  choice = int(input("\n1. Addition\n2. Substraction\n3. Multiplaction\n4. Division\n5. Square\n6. Cube\n7. Exit\n\nEnter your choice(1-7): "))

  match choice:
    case 1:
      print("---------- Addition ----------")
      n1, n2 = int(input("n1 = ")), int(input("n2 = "))
      print(f"{n1} + {n2} = {(add(n1,n2))} ")

    case 2:
      print("---------- Substraction ----------")
      n1, n2 = int(input("n1 = ")), int(input("n2 = "))
      print(f"{n1} - {n2} = {(sub(n1,n2))} ")

    case 3:
      print("---------- Multiplaction ----------")
      n1, n2 = int(input("n1 = ")), int(input("n2 = "))
      print(f"{n1} x {n2} = {(mul(n1,n2))} ")

    case 4:
      print("---------- Division ----------")
      n1, n2 = int(input("n1 = ")), int(input("n2 = "))
      print(f"{n1} / {n2} = {(div(n1,n2))} ")

    case 5:
      print("---------- Square ----------")
      n = int(input("Enter a number: "))
      print(f"Square of {n} =  {(square(n))} ")

    case 6:
      print("---------- Cube ----------")
      n = int(input("Enter a number: "))
      print(f"Cube of {n} =  {(cube(n))} ")

    case 7:
      print("Exiting...") 
      exit()
    case _:
      print("Invalid choice !")  
    
```

#### 16. build_sandwich(*ingredients) function.


```python
# write your code here 
def build_sandwich(*ingredients) :
  print("Ingredients required prepare sandwich are:")
  num = 1
  for e in ingredients :
    print(f"{num}. {e}")
    num += 1

ingredients = ['Bun', 'Potato', 'Butter', 'Salad', 'Soace']
build_sandwich(*ingredients)
```

#### 18. Count *args and **kwargs in one function.


```python
# write your code here 
def count_arg(*args, **kwargs) :
  arg_count = 0
  kwarg_count = 0
  for e in args :
    arg_count += 1

  for e in kwargs :
    kwarg_count += 1
  print(f"Argument count = {arg_count}")
  print(f"Keyword Argument count = {kwarg_count}")

l = [10,20,30,40]
d = {'Name': 'Anshuman', 'Age': 22, 'Roll_num': 57}
count_arg(*l, **d)
```
